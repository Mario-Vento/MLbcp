```python
# ============================================================
# Compute : Edad
# ============================================================

df_to_scoring = df_to_scoring.withColumn(
    "ref_date",
    F.last_day(F.to_date(F.concat_ws("-", F.col("codmes").substr(1,4), F.col("codmes").substr(5,2), F.lit("01"))))
).withColumn(
    "edad",
    F.when(F.col("dem_fec_nacimiento").isNull(), F.lit(None).cast("int"))
    .when(F.col("dem_fec_nacimiento") > F.col("ref_date"), F.lit(None).cast("int"))
    .otherwise(F.floor(F.months_between(F.col("ref_date"), F.col("dem_fec_nacimiento")) / 12))
).drop("dem_fec_nacimiento", "ref_date")
```

```python
# ============================================================
# Compute : max_mora_intra_u6m
# ============================================================

# F.greatest() ya excluye nulls automáticamente

df_to_scoring = (
    df_to_scoring
    .withColumn(
        "max_mora_intra_u3m",
        F.greatest(
            F.col("ctdmora_intra_0"),
            F.col("ctdmora_intra_1"),
            F.col("ctdmora_intra_2")
        )
    )
    .withColumn(
        "max_mora_intra_p3m",
        F.greatest(
            F.col("ctdmora_intra_3"),
            F.col("ctdmora_intra_4"),
            F.col("ctdmora_intra_5")
        )
    )
    .withColumn(
        "max_mora_intra_u6m",
        F.greatest(
            F.col("max_mora_intra_u3m"),
            F.col("max_mora_intra_p3m")
        )
    )
)
```

```python
# ============================================================
# Compute : max_mora_intra_g3m
# ============================================================

A = col("max_mora_intra_u3m")  # Numerador
B = col("max_mora_intra_p3m")  # Denominador

df_to_scoring = (
    df_to_scoring
    .withColumn(
        "max_mora_intra_g3m",
        (
            # === A es NULL ===
            when(A.isNull() & B.isNull(), lit(44444444444))
            .when(A.isNull() & (B == 0), lit(6666666666))
            .when(A.isNull() & (B > 0), lit(1111111111))
            .when(A.isNull() & (B < 0), lit(-1111111111))
            # === A == 0 ===
            .when((A == 0) & B.isNull(), lit(7777777777))
            .when((A == 0) & (B == 0), lit(5555555555))
            # (A==0 & B!=0) cae a la división -> 0.0
            # === B es NULL (A no null, A != 0) ===
            .when(B.isNull() & (A > 0), lit(2222222222))
            .when(B.isNull() & (A < 0), lit(-2222222222))
            # === B == 0 (A no null, A != 0) ===
            .when((B == 0) & (A > 0), lit(3333333333))
            .when((B == 0) & (A < 0), lit(-3333333333))
            # === División válida ===
            .otherwise(round(A / B, 8))
        )
        .cast(DecimalType(19, 8))
    )
)
```

```python
# ============================================================
# Compute : flg_titulo
# ============================================================

df_to_scoring = (
    df_to_scoring
    # flg_titulo = "1" si cualquiera de los flags = "1", si no "0"
    .withColumn(
        "flg_titulo",
        F.when(
            (F.col("flgniveleducativobachiller").cast("string") == F.lit("1")) |
            (F.col("flgniveleducativodoctorado").cast("string") == F.lit("1")) |
            (F.col("flgniveleducativomaestria").cast("string") == F.lit("1")) |
            (F.col("flgniveleducativotitulado").cast("string") == F.lit("1")),
            F.lit("1")
        ).otherwise(F.lit("0"))
    )
)
```

```python
# ============================================================
# Compute : rcc_mto_deu_ship_max_u12_u24 : rcc_mto_deu_ship_max_g12m
# ============================================================

A = col("rcc_mto_deu_ship_max_u12")  # Numerador
B = col("rcc_mto_deu_ship_max_u24")  # Denominador

df_to_scoring = (
    df_to_scoring
    .withColumn(
        "rcc_mto_deu_ship_max_u12_u24",
        (
            # === A es NULL ===
            when(A.isNull() & B.isNull(), lit(44444444444))
            .when(A.isNull() & (B == 0), lit(6666666666))
            .when(A.isNull() & (B > 0), lit(1111111111))
            .when(A.isNull() & (B < 0), lit(-1111111111))
            # === A == 0 ===
            .when((A == 0) & B.isNull(), lit(7777777777))
            .when((A == 0) & (B == 0), lit(5555555555))
            # (A==0 & B!=0) cae a la división -> 0.0
            # === B es NULL (A no null, A != 0) ===
            .when(B.isNull() & (A > 0), lit(2222222222))
            .when(B.isNull() & (A < 0), lit(-2222222222))
            # === B == 0 (A no null, A != 0) ===
            .when((B == 0) & (A > 0), lit(3333333333))
            .when((B == 0) & (A < 0), lit(-3333333333))
            # === División válida ===
            .otherwise(round(A / B, 8))
        )
        .cast(DecimalType(19, 8))
    )
)
```
