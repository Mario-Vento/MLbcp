#Variables Implementación Troncal Cliente

**1) ctdmora_intra_0_o**
Variable de atraso intramés (UM) que captura el nivel de mora registrado dentro del mes vigente. Se utiliza como señal temprana de deterioro, al reflejar incumplimientos recientes en el comportamiento de pagos. Se encuentra capado a 15. Corresponde a la familia ATRASO.

**2) max_maduracion_cli**
Mide la maduración/antigüedad del cliente (experiencia), usada para diferenciar perfiles nuevos vs maduros. En el diccionario aparece como “maduración cliente” dentro de la familia EXPERIENCIA y se construye desde el portafolio de crédito.

**3) max_mora_intra_u6m_o**
Representa el máximo atraso intramés observado en la ventana de últimos 6 meses (U6M). Su objetivo es capturar picos de incumplimiento recientes, aun cuando el cliente haya retornado a un estado de no atraso. Pertenece a la familia ATRASO y se origina en la lógica “intra”.

**4) exp_pct_evol_ship_u3m_000**
Variable de evolución de deuda SHIP a nivel trimestral, construida como un ratio (RT) que compara el comportamiento reciente (U3M) versus una referencia (U6M). Se usa para identificar incrementos/disminuciones relativas del saldo/endeudamiento SHIP en el corto plazo. Pertenece a la familia DEUDA.

**5) prd_pct_pmpas_pmact_2_000_o**
Ratio pasivos/activos en ventana U24M, calculado como relación entre promedios (pasivos vs activos). Se interpreta como una aproximación a la estructura financiera y nivel de apalancamiento del cliente en el mediano plazo. Está en la familia PASIVOS.

**6) fatc_pct_pag_mn_ctami_000_o**
Ratio entre pago mínimo y deuda promedio de tarjeta de crédito en U6M. Esta variable resume el patrón de pago (dependencia del mínimo vs amortización real), útil para diferenciar clientes que tienden a “rodar” deuda. Está en la familia TRX TC.

**7) rcc_pct_sf3_sf24_ship_000**
Ratio de deuda SHIP (Total de deuda sin considerar créditos hipotecarios) entre U3M y U24M, que mide qué tan representativo es el endeudamiento reciente frente al nivel estructural de 24 meses. Se usa para detectar aceleraciones en la deuda SHIP. Está en la familia DEUDA.

**8) rcc_pct_rdv_prm_u3m_ooo**
Variable asociada al porcentaje promedio de devengados en U3M dentro del entorno RCC (devengados como métrica definida por la fuente). Se usa para resumir dinámica reciente de devengamiento relacionada a deuda/exposición. Pertenece a la familia DEUDA.

**9) rcc_q_mes_act_sf_buen_000**
Indicador “Bueno menos Malo” en RCC para U3M, entendido como balance neto de meses con condición favorable versus desfavorable dentro del sistema financiero. Se usa para aproximar estabilidad reciente de comportamiento crediticio agregado. En tu diccionario se clasifica en ATRASO.

**10) prd_prm_tsav_mnn_6_6_rt6_ooo**
Ratio entre ahorro mínimo y ahorro promedio en U6M, usado para medir estabilidad (o variabilidad) del ahorro: valores bajos sugieren mayor volatilidad o episodios de saldo mínimo muy reducido. Se ubica en la familia de PASIVOS.

**11) flg_titulo**
Indicador binario que señala si el cliente registra un nivel educativo superior a bachiller. Se usa como proxy socio-demográfica (capital humano) en el ordenamiento del riesgo/comportamiento. En el diccionario pertenece a la familia DEMOGRAFICO.

**12) q_diamora_max_100_u24_o**
Máximo atraso observado en U24M bajo la definición/umbral “100”. Captura la severidad histórica del atraso, priorizando el peor evento dentro de la ventana. Está en la familia ATRASO.

**13) fatc_flg_pag_ful_clan_000**
Indicador (flag) relacionado a abono total de deuda de tarjeta en U3M, utilizado para capturar comportamiento de pago “full” o cancelación en el corto plazo. Pertenece a la familia TRX TC.

**14) rcc_pct_sf12_sf24_rt_u24**
Ratio de deuda SHIP/SSFF entre U12M y U24M, orientado a capturar cambios de mediano plazo respecto a la base estructural de 24 meses. Pertenece a la familia de DEUDA.

**15) edad_o**
Edad del cliente derivada de la fecha de nacimiento registrada. Se utiliza como variable demográfica base para capturar diferencias estructurales de comportamiento asociadas al ciclo de vida. Pertenece a la familia DEMOGRAFICO.

**16) pos_pct_q_etcnpscl_a__000**
Variable de consumo POS asociada a compras en establecimientos “Top”, agregada en ventana U6M. Pertenece a la familia TRX.

**17) ctdpdhu24_ooo**
Cuenta la cantidad de abonos en cuenta sueldo (PDH) en U24M, como señal de recurrencia de ingresos por planilla o estabilidad laboral. Pertenece a la familia PDH.

**18) grf_tip_clas_rie_cli__000**
Variable de grafos que refleja el % de clientes cercanos cuya clasificación de riesgo se encuentra en rangos 2 a 4 (asociado a perfiles “> DEF” en la descripción), en ventana U3M. Se interpreta como señal de “contagio” o correlación por vecindad/relación. Pertenece a la familia GRAFO.

**19) isav_q_opea_desm_prm_u3m_o**
Cuenta la cantidad promedio de abonos vinculados a desembolsos en U3M. Se interpreta como una señal de actividad asociada a ingresos/desembolsos o movimientos recurrentes de entrada. Esta en la familia de TRX.

**20) grf_cvta_prov_rie4_pr_000**
Variable de grafo que mide un % (ponderado/por ventas) asociado a proveedores cercanos con clasificación de riesgo alta (rie4) en U3M. Se usa como señal contextual del entorno relacional del cliente. Pertenece a la familia GRAFO.

**21) prod_flg_sld_aho_300**
Indicador que marca si el cliente presenta saldo de ahorro mayor a 300 bajo el concepto de resumen saldo. Se usa como señal de colchón de liquidez o tenencia mínima de ahorro. Pertenece a la familia de PASIVOS.

**22) rcc_mto_deu_ship_max__000**
Ratio de evolución de deuda SHIP que compara el componente U12M frente a U24M (“U12M/U24M”), orientado a capturar cambios estructurales del endeudamiento SHIP. Pertenece a la familia DEUDA.

**23) max_mora_intra_g3m**
Variable de evolución de mora intramés en un horizonte trimestral (G3M según naming). Resume cambios recientes en la mora intramés y se utiliza para capturar tendencias (mejora/deterioro) de corto plazo. En el diccionario está dentro de ATRASO y lógica “intra”.

**24) q_mes_mto_tot_pgsrv_s_000**
Cuenta la cantidad de veces/meses en que el cliente realizó pagos de servicios (agua, luz, etc.) con monto > 0 en U3M. Se usa como señal transaccional de recurrencia/actividad de pago de servicios. Pertenece a la familia SERVICIOS.
